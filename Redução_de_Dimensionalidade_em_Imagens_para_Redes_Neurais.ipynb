**Binarização** - Convertendo imagens em tons de cinza para imagens preto e branco onde todos os pixels possuem 0 ou 255 como valores de intensidade.

### Importando Imagem para o nosso trabalho

```python
img = cv2.imread("/content/drive/MyDrive/fotos/maria.jpg")
cv2_imshow(img)
cv2.waitKey(0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
suave = cv2.GaussianBlur(img, (7, 7), 0)
(T, bin) = cv2.threshold(suave, 161, 255, cv2.THRESH_BINARY)
(T, binI) = cv2.threshold(suave, 161, 255, cv2.THRESH_BINARY_INV)
resultado = np.vstack([
    np.hstack([suave, bin]),
    np.hstack([binI, cv2.bitwise_and(img, img, mask = binI)])
])
cv2_imshow(resultado)
cv2.waitKey(0)
```
